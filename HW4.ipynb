1. Расскажите о том как работает регуляризация в решающих деревья, какие параметры мы штрафуем в данных алгоритмах?

Регуляризация позволяет уменьшать размер дерева решений, удаляя участки дерева, которые имеют маленький вес.

Важен оптимальный размер конечного дерева. Так, небольшое дерево может не охватить ту или иную важную информацию о выборочном пространстве. Общая стратегия ограничения дерева это удаление узлов в случае, если они не дают дополнительной информации.
Сокращение дерева может осуществляться сверху вниз или снизу вверх. Сверху вниз — обрезка начинается с корня, снизу вверх — сокращается число листьев дерева. Один из простейших методов регулирования — уменьшение ошибки ограничения дерева. Начиная с листьев, каждый узел заменяется на самый популярный класс. Если изменение не влияет на точность предсказания, то оно сохраняется.
В зависимости от библиотеки применяемой для обучения модели, параметрами для штрафа являются параметры соответствующие коэффициентам  для L1-регуляризации и L2-регуляризации.

2. По какому принципу рассчитывается "важность признака (feature_importance)" в ансаблях деревьев?

Важность признака возвращает вектор "важностей" признаков. Индекс элемента в этом векторе соответствует индексу признака в данных. Значение элемента отражает "важность" признака относительно остальных: чем больше значение, тем больше важность.
