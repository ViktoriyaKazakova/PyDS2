1. Привести по 2 примера, когда лучше максимизировать Precision, а когда Recall?

Precision (точность) - показывает, сколько из предсказанных положительных примеров оказались действительно положительными. Чем больше точность, тем меньше число ложно положительных попаданий. Таким образом, стоит максимизировать эту метрику в тех случаях, когда каждое дополнительное ложное попадание черевато нецелесообразными дополнительными финансовыми затратами или потерями для бизнеса, например:

ложное определение привело к лишним выплатам по страховым обязательствам;
ложное определение привело к избыточному складскому запасу и "затовариванию";

Recall (полнота) - показывает, сколько от общего числа фактических положительных примеров было предсказано как положительный класс. Мера полноты характеризует способность классификатора определять как можно большее число положительных ответов из ожидаемых. Таким образом, стоит максимизировать эту метрику в тех случаях, когда пропуск объекта целевого класса череват созданием увеличением рисков различных форм и последствий. Т.е. мы максимизируем данную метрику, когда нам необходимо определить все положительные примеры и когда важно снизить количество ложно отрицательных примеров, например:

важно выявить всех пациентов больных каким-либо заболеванием, при этом возможно включив в это число здоровых пациентов и позже отправив их на дообследование;
важно выявить все бракованные детали в процессе изготовления, до этапа сборки и эксплуатации, при этом возможно включив в это число небракованные детали и отправив их вместе с браком на переработку;

2. Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?

Итак, хотя точность и полнота являются очень важными метриками, сами по себе они не дадут нам полной картины и поэтому мы используем метрику F-score, которая позволяет объединить в себе значения метрик точность и полнота. А коэффициентт даёт возможность балансировать, тем самым позволяя влиять на значимость той метрики, максимизация которой для нас наиболее приоритетна.
